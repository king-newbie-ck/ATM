```python
# ─────────────────────────────────────────────
#  ATM Simulation Program
# ─────────────────────────────────────────────

# Account Database
accounts = {
    "50010101234": {"pin": "1111", "name": "Avinash",  "balance": 125000.00},
    "50010104321": {"pin": "2222", "name": "Sangeeth", "balance": 750000.00},
}

MAX_PIN_ATTEMPTS = 3


# ── Login ─────────────────────────────────────
def login():
    print("\n" + "=" * 40)
    print("        Welcome to Python ATM")
    print("=" * 40)

    acc_no = input("Enter Account Number : ").strip()

    if acc_no not in accounts:
        print("❌ Account not found.")
        return None

    for attempt in range(1, MAX_PIN_ATTEMPTS + 1):
        pin = input(f"Enter PIN (Attempt {attempt}/{MAX_PIN_ATTEMPTS}) : ").strip()
        if pin == accounts[acc_no]["pin"]:
            print(f"\n✅ Welcome, {accounts[acc_no]['name']}!")
            return acc_no
        else:
            print("❌ Wrong PIN.")

    print("🔒 Card blocked after 3 wrong attempts.")
    return None


# ── Check Balance ─────────────────────────────
def check_balance(acc_no):
    balance = accounts[acc_no]["balance"]
    print(f"\n💰 Available Balance : ₹ {balance:,.2f}")


# ── Deposit ───────────────────────────────────
def deposit(acc_no):
    try:
        amount = float(input("\nEnter amount to deposit : ₹ "))
        if amount <= 0:
            print("❌ Amount must be greater than 0.")
            return
        accounts[acc_no]["balance"] += amount
        print(f"✅ ₹ {amount:,.2f} deposited successfully.")
        print(f"💰 New Balance : ₹ {accounts[acc_no]['balance']:,.2f}")
    except ValueError:
        print("❌ Invalid amount.")


# ── Withdraw ──────────────────────────────────
def withdraw(acc_no):
    try:
        amount = float(input("\nEnter amount to withdraw : ₹ "))
        if amount <= 0:
            print("❌ Amount must be greater than 0.")
        elif amount > accounts[acc_no]["balance"]:
            print("❌ Insufficient balance.")
        else:
            accounts[acc_no]["balance"] -= amount
            print(f"✅ ₹ {amount:,.2f} withdrawn successfully.")
            print(f"💰 Remaining Balance : ₹ {accounts[acc_no]['balance']:,.2f}")
    except ValueError:
        print("❌ Invalid amount.")


# ── Change PIN ────────────────────────────────
def change_pin(acc_no):
    old_pin = input("\nEnter current PIN : ").strip()
    if old_pin != accounts[acc_no]["pin"]:
        print("❌ Wrong PIN. Cannot change.")
        return
    new_pin = input("Enter new PIN      : ").strip()
    if not new_pin.isdigit() or len(new_pin) != 4:
        print("❌ PIN must be 4 digits.")
        return
    confirm = input("Confirm new PIN    : ").strip()
    if new_pin != confirm:
        print("❌ PINs do not match.")
        return
    accounts[acc_no]["pin"] = new_pin
    print("✅ PIN changed successfully.")


# ── Main Menu ─────────────────────────────────
def menu(acc_no):
    while True:
        print("\n" + "-" * 40)
        print("         ATM MENU")
        print("-" * 40)
        print("  1. Check Balance")
        print("  2. Deposit")
        print("  3. Withdraw")
        print("  4. Change PIN")
        print("  5. Exit")
        print("-" * 40)

        choice = input("Choose an option (1-5) : ").strip()

        if choice == "1":
            check_balance(acc_no)
        elif choice == "2":
            deposit(acc_no)
        elif choice == "3":
            withdraw(acc_no)
        elif choice == "4":
            change_pin(acc_no)
        elif choice == "5":
            print("\n✅ Thank you for using Python ATM. Goodbye!")
            break
        else:
            print("❌ Invalid choice. Enter 1 to 5.")


# ── Run ───────────────────────────────────────
acc_no = login()
if acc_no:
    menu(acc_no)
```
